```python
import time
import os
from IPython.display import Video, display

# Function to measure inference time
def run_inference(script_path, label):
    print(f"\nRunning {label} inference...")
    start = time.time()
    exit_code = os.system(f"python {script_path}")
    end = time.time()

    if exit_code == 0:
        print(f"{label} inference finished in {end - start:.2f} seconds.")
    else:
        print(f"{label} inference failed with exit code {exit_code}.")
    return end - start

# Paths
original_script = "inference.py"
optimized_script = "inference_optimized.py"

# Run both and record times
original_time = run_inference(original_script, "Original")
optimized_time = run_inference(optimized_script, "Optimized")

# Compare inference times
print("\nâ± Inference Time Comparison:")
print(f"Original  : {original_time:.2f} seconds")
print(f"Optimized : {optimized_time:.2f} seconds")
improvement = original_time - optimized_time
if improvement > 0:
    print(f"âœ… Optimization saved {improvement:.2f} seconds!")
else:
    print(f"âš ï¸ No improvement. Optimization slower by {abs(improvement):.2f} seconds.")

# Video display
original_video_path = "animations/s0--d0_concat.mp4"
optimized_video_path = "animations/s0--d0_concat_optimized.mp4"

print("\nðŸŽ¥ Original Video:")
if os.path.exists(original_video_path):
    display(Video(original_video_path, embed=True, width=320))
else:
    print("âŒ Not found:", original_video_path)

print("\nðŸŽ¥ Optimized Video:")
if os.path.exists(optimized_video_path):
    display(Video(optimized_video_path, embed=True, width=320))
else:
    print("âŒ Not found:", optimized_video_path)
```
